```python
import random
import tkinter as tk
from tkinter import messagebox

# Function to handle the game logic
def play_game(user_choice):
    global user_count, comp_count, games_played

    # Mapping of user's input and random computer input
    comp_choice = random.choice(list(my_dict.keys()))

    # Display choices
    comp_choice_label.config(text=f"Computer's Choice: {my_dict[comp_choice]}")
    user_choice_label.config(text=f"Your Choice: {my_dict[user_choice]}")

    # Determine the winner
    if (user_choice == 'R' and comp_choice == 'P') or (user_choice == 'P' and comp_choice == 'S') or (
            user_choice == 'S' and comp_choice == 'R'):
        comp_count += 1
        result_label.config(text="Computer Wins!", fg="red")
    elif (user_choice == 'P' and comp_choice == 'R') or (user_choice == 'S' and comp_choice == 'P') or (
            user_choice == 'R' and comp_choice == 'S'):
        user_count += 1
        result_label.config(text="You Win!", fg="green")
    else:
        result_label.config(text="It's a Tie!", fg="blue")

    # Update the score labels
    user_score_label.config(text=f"Your Score: {user_count}")
    comp_score_label.config(text=f"Computer's Score: {comp_count}")
    
    games_played += 1

    # Check if the game has reached the set number of rounds
    if games_played >= total_games:
        show_final_result()

def show_final_result():
    if user_count > comp_count:
        messagebox.showinfo("Game Over", "Congratulations! You won the game!")
    elif user_count < comp_count:
        messagebox.showinfo("Game Over", "Sorry, you lost the game!")
    else:
        messagebox.showinfo("Game Over", "It's a tie!")

    # Disable buttons after the game ends
    rock_button.config(state="disabled")
    paper_button.config(state="disabled")
    scissors_button.config(state="disabled")

def start_game():
    global total_games, user_count, comp_count, games_played
    total_games = int(game_entry.get())
    user_count = 0
    comp_count = 0
    games_played = 0

    # Reset scores and labels
    user_score_label.config(text="Your Score: 0")
    comp_score_label.config(text="Computer's Score: 0")
    result_label.config(text="")
    comp_choice_label.config(text="")
    user_choice_label.config(text="")
    
    # Enable buttons for gameplay
    rock_button.config(state="normal")
    paper_button.config(state="normal")
    scissors_button.config(state="normal")

# Create the main window
root = tk.Tk()
root.title("Rock Paper Scissors Game")
root.geometry("400x400")

# Apply eye-comfortable colors
bg_color = "#e1f0e8"  # Soft light green background
btn_color = "#a8d5ba"  # Softer green for buttons
text_color = "#2e4f4f"  # Dark green for text
highlight_color = "#f6e58d"  # Soft yellow for highlighted areas

root.configure(bg=bg_color)

# Variables
my_dict = {'R': "Rock", 'P': "Paper", 'S': "Scissors"}
user_count = 0
comp_count = 0
games_played = 0
total_games = 0

# Widgets for the game
game_label = tk.Label(root, text="Enter Number of Rounds:", bg=bg_color, fg=text_color, font=("Helvetica", 12))
game_label.pack(pady=10)

game_entry = tk.Entry(root, bg=highlight_color, fg=text_color)
game_entry.pack(pady=5)

start_button = tk.Button(root, text="Start Game", bg=btn_color, fg=text_color, command=start_game)
start_button.pack(pady=10)

# Choice buttons
rock_button = tk.Button(root, text="Rock", bg=btn_color, fg=text_color, width=15, state="disabled", command=lambda: play_game('R'))
rock_button.pack(pady=5)

paper_button = tk.Button(root, text="Paper", bg=btn_color, fg=text_color, width=15, state="disabled", command=lambda: play_game('P'))
paper_button.pack(pady=5)

scissors_button = tk.Button(root, text="Scissors", bg=btn_color, fg=text_color, width=15, state="disabled", command=lambda: play_game('S'))
scissors_button.pack(pady=5)

# Displaying Results
user_choice_label = tk.Label(root, text="", bg=bg_color, fg=text_color, font=("Helvetica", 10))
user_choice_label.pack(pady=5)

comp_choice_label = tk.Label(root, text="", bg=bg_color, fg=text_color, font=("Helvetica", 10))
comp_choice_label.pack(pady=5)

result_label = tk.Label(root, text="", font=("Helvetica", 14), bg=bg_color)
result_label.pack(pady=10)

# Displaying the scores
user_score_label = tk.Label(root, text="Your Score: 0", bg=bg_color, fg=text_color, font=("Helvetica", 12))
user_score_label.pack()

comp_score_label = tk.Label(root, text="Computer's Score: 0", bg=bg_color, fg=text_color, font=("Helvetica", 12))
comp_score_label.pack()

# Start the main loop
root.mainloop()
```
